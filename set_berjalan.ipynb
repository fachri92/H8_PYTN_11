```python
def set_map(x):
    if x == 'TOTAL_NET_CLAIM_AMOUNT' or x=='TOTAL_CLAIM' or x=='PurePremium' or x=='actual_loss_ratio':
        return 'ACTUAL'
    elif x=='freq_pred' or x=='sev_pred' or  x=='ExpectedLoss' or x=='predicted_loss_ratio' :
        return 'PREDICTED'
    else:
        return 'RATIO'
```

```python
def set_map2(x):
    if x == 'TOTAL_NET_CLAIM_AMOUNT' or  x=='ExpectedLoss' or x=='loss_ratio' :
        return 'LOSS'
    elif x=='freq_pred'or x=='TOTAL_CLAIM' or x=='freq_ratio' :
        return 'FREQUENCY'
    elif x=='actual_loss_ratio' or x=='predicted_loss_ratio':
        return 'LOSS RATIO'
    else:
        return 'SEVERITY'
```

```python
set_berjalan_concat_melted['ACTUAL_PREDICTED_RATIO'] = set_berjalan_concat_melted.variable.apply(set_map)
```

```python
set_berjalan_concat_melted['SEVERITY_FREQUENCY_LOSS']= set_berjalan_concat_melted.variable.apply(set_map2)
```
